```python
import numpy as np
import matplotlib.pyplot as plt

# intervalo de tempo entre amostras
deltaT = 0.0001
T0 = 0
Tf = 0.02
Tempo = np.arange(T0, Tf, deltaT)
Xt = np.cos(100 * np.pi * Tempo)
Ns = [4, 6, 8, 10]  # Valores de N para teste
errors = []

for N in Ns:
    mi = np.min(Xt)
    Xi = Xt - mi
    ma = np.max(Xi)
    Xnorm = Xi / ma

    Yt = ma * np.around(Xt * (2 ** N - 1)) / (2 ** N - 1)
    Yt = Yt + np.round(mi)

    abs_error = np.abs(Xt - Yt)
    errors.append(abs_error)

    plt.figure()

    plt.plot(Tempo, Xt, 'b')  # plotando sinal cont√≠nuo
    plt.plot(Tempo, Yt, 'r')

    plt.stem(Tempo, Xt, 'g')  # plotando sinal discreto
    plt.stem(Tempo, Yt, 'k')
    plt.title(f'N = {N}')
    plt.show()

plt.figure()
for i, N in enumerate(Ns):
    plt.plot(Tempo, errors[i], label=f'N = {N}')

plt.xlabel('Tempo')
plt.ylabel('Erro Absoluto')
plt.legend()
plt.show()
```
